```python
from skimage import data,color,io
from PIL import Image
from matplotlib import pyplot as plt
```

```python
image = data.coffee()
plt.imshow(image)
```

# **Convert into Grayscale Image**

```python
gray_image = color.rgb2gray(image)
plt.imshow(gray_image)
```

# **Rescale Image**

```python
from skimage.transform import rescale, resize, downscale_local_mean
```

```python
rescale_image = rescale(gray_image, 3.0/4.0, anti_aliasing = True)
plt.imshow(rescale_image)
```

# **Resize Image**

```python
resize_image = resize(image, (200,200))
plt.imshow(resize_image)
```

# **Downscale Image**

```python
downscale_image = downscale_local_mean(gray_image, (10,10))
plt.imshow(downscale_image)
```

# **Image Filtering**

```python
from skimage.filters import roberts, sobel, scharr, prewitt
```

```python
edge_roberts = roberts(gray_image)
plt.imshow(edge_roberts, cmap = 'gray')
```

```python
edge_sobel = sobel(gray_image)
plt.imshow(edge_sobel, cmap = 'gray')
```

```python
edge_scharr = scharr(gray_image)
plt.imshow(edge_scharr, cmap='gray')
```

# **Canny Edge Detection**

```python
from skimage.feature import canny
```

```python
edge_canny = canny(gray_image, sigma = 3)
plt.imshow(edge_canny, cmap= 'gray') 
```

# **Convolution Image**

```python
import numpy as np
from skimage import restoration
```

```python
psf = np.ones((3,3)) / 9
```

```python
deconvolved, _ = restoration.unsupervised_wiener(gray_image, psf)
plt.imshow(deconvolved, cmap = 'gray')
```
