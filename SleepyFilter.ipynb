# **Extract features from a person image to differentiate whether the person is sleepy or not using image processing**

```python
from PIL import Image
from skimage import data, color, io, img_as_float
import matplotlib.pyplot as plt
from scipy import ndimage as nd
```

```python
image = io.imread('/content/image1357.jpg')
plt.imshow(image)
```

**Crop Image**


```python
x1 = 550
x2 = 750
y1 = 250
y2 = 750
cropped = image[x1:x2,y1:y2]
plt.imshow(cropped)
```

**Change to grayscale**

```python
gray_image = color.rgb2gray(cropped)
plt.imshow(gray_image, cmap='gray')
```

```python
from skimage.feature import canny
from skimage.filters import roberts, sobel, scharr, prewitt
```

**Canny Edge Detection**

```python
edge_image = canny(gray_image, sigma = 1)
plt.imshow(edge_image, cmap='gray')
```

```python
edge_roberts = roberts(gray_image)
plt.imshow(edge_roberts, cmap='gray')
```

```python
edge_sobel = sobel(gray_image)
plt.imshow(edge_sobel, cmap = 'gray')
```
