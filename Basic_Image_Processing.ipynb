# **Basic Image Processing using Scikit-Image Library**

```python
from matplotlib import pyplot as plt
from skimage import io,color
from skimage.transform import rescale, resize, downscale_local_mean
```

# **Read Image**

```python
# read sample image
read_image = io.imread("/SampleCell.jpg", as_gray = True)
```

```python
plt.imshow(read_image)
```

# **Rescale Image**


```python
# rescale image
rescale_image = rescale(read_image, 1.0 / 4.0, anti_aliasing=False)
```

```python
plt.imshow(rescale_image)
```

# **Resize Image**

```python
# resize image
resize_image = resize(read_image, (80, 80), anti_aliasing=True)
```

```python
plt.imshow(resize_image)
```

```python
plt.imshow(resize_image, cmap='gray')
```
