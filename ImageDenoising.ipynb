# **Denoising microscope image in Python**

```python
from skimage import io, img_as_float
from matplotlib import pyplot as plt
from scipy import ndimage as nd

import numpy as np
```

# **Gaussian Filter**

```python
img = io.imread("/content/cell1.jpg")
plt.imshow(img)
```

```python
gaussian_image = nd.gaussian_filter(img, sigma = 5)
plt.imshow(gaussian_image)
```

# **Median Filter**

```python
median_image = nd.median_filter(img, size=3)
plt.imshow(median_image)
```

# **Non-Local Means Denoising**

```python
from skimage.restoration import denoise_nl_means, estimate_sigma
```

```python
float_image = img_as_float(img)
```

```python
sigma_est = np.mean(estimate_sigma(float_image, multichannel=True))
denoise_image = denoise_nl_means(float_image, h=1.15*sigma_est, fast_mode=True, patch_size=3, patch_distance=6,multichannel=True)
plt.imshow(denoise_image)
```
